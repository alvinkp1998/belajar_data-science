# Add Column Drop Column

```python
#add column
df['try'] = 'Try'
#delete column
df.drop(columns='try', inplace=True)
```

```python
#bikin kolom baru
df['new_column'] = df['calories'] * df['protein']
```

```python
df.drop(columns='new_column', inplace=True)
```

## Map, Apply, and Transform

### Map

```python
posisi = {
    1: 'bawah',
    2: 'tengah',
    3: 'atas'
}

df['shelf_name'] = df['shelf'].map(posisi)
```

```python
df.drop(columns='shelf', inplace = True)
```

```python
df.head()
```

## Apply

```python
from luwiji.pandas import illustration
```

```python
def g_to_mg(X):
    return X*1000
```

```python
df['fat_in_mg'] = df['fat'].apply(g_to_mg)
df.head()
```

```python
df['protein_in_mg'] = df['protein'].apply(lambda X: X*1000)
df.head()
```

```python
type(df['vitamins'][0])
```

```python
df['vitamins'] = df['vitamins'].apply(lambda X: X.lower())
df.head()
```

### Transform

```python
df['proporsi_rating'] = df['rating'].transform(lambda X: X/X.sum())
```

```python
df.head()
```
