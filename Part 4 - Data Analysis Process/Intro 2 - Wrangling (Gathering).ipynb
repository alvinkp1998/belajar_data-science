```python
import pandas as pd
```

```python
df = pd.read_csv('./data/student_scores.csv')
df.head()
```

```python
# melakukan iterasi untuk setiap nama kolom pada dataframe, merubah nama kolom menjadi lowercase, dan
# disimpan ke dalam sebuah variabel baru dalam bentuk list
new_col = []
for col in df.columns:
    new_col.append(col.lower())
```

```python
# merubah nama kolom yang ada dengan nama kolom baru (yang sudah di lower case)
df.columns = new_col
```

```python
df.head()
```

```python
# menjadikan kolom id sebagai index dan overwrite variabel df
df.set_index('id', inplace = True)
```
